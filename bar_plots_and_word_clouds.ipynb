```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
```

## Barplots by category (sum of views)

```python
def process_data(country, column):
    df =  pd.read_csv("csv_files/" + country + '_videos.csv')[['trending_date', 'title', 'channel_title', 'category','tags', 
                                                'description', 'views', 'likes', 'dislikes', 'comment_count']]
    grouped_df = df.groupby('category').sum()
    column_dict = grouped_df[column].to_dict()
    return column_dict
```

```python
gb_views = process_data('gb', 'views')
us_views = process_data('us', 'views')
ca_views = process_data('ca', 'views')
de_views = process_data('de', 'views')
fr_views = process_data('fr', 'views')
```

```python
trace1 = go.Bar(
    x=gb_views.values(),
    y=gb_views.keys(),
    name='gb',
    orientation = 'h'
)

trace2 = go.Bar(
    x=us_views.values(),
    y=us_views.keys(),
    name='us',
    orientation = 'h'
)

trace3 = go.Bar(
    x=ca_views.values(),
    y=ca_views.keys(),
    name='ca',
    orientation = 'h'
)

trace4 = go.Bar(
    x=de_views.values(),
    y=de_views.keys(),
    name='de',
    orientation = 'h'
)

trace5 = go.Bar(
    x=fr_views.values(),
    y=fr_views.keys(),
    name='fr',
    orientation = 'h'
)

data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

## Boxplot

```python
gb = pd.read_csv('csv_files/gb_videos.csv')
us = pd.read_csv('csv_files/us_videos.csv')
ca = pd.read_csv('csv_files/ca_videos.csv')
de = pd.read_csv('csv_files/de_videos.csv')
fr = pd.read_csv('csv_files/fr_videos.csv')

trace0 = go.Box(
    y=gb['views']
)
trace1 = go.Box(
    y=us['views']
)
trace2 = go.Box(
    y=ca['views']
)
trace3 = go.Box(
    y=de['views']
)
trace4 = go.Box(
    y=fr['views']
)


data = [trace0, trace1, trace2, trace3, trace4]
py.iplot(data)
```

## Word Cloud (title)

```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

#mpl.rcParams['figure.figsize']=(8.0,6.0)    #(6.0,4.0)
mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=100             #72 
mpl.rcParams['figure.subplot.bottom']=.1 


stopwords = set(STOPWORDS)
data = gb

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(data['title']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```
